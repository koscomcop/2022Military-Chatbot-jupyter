```python
def trim_author(author):

    if not isinstance(author, str):
        return

    if author[-5:] == ' 글·그림':
        author = author[:-5]
    if author[-5:] == ' 글·사진':
        author = author[:-5]
    if author[-3:] == ' 지음':
        author = author[:-3]
    if author[-3:] == ' 편저':
        author = author[:-3]
    if author[-3:] == ' 엮음':
        author = author[:-3]
    if author[-3:] == ' 옮김':
        author = author[:-3]
    if author[-3:] == ' 역주':
        author = author[:-3]
    if author[-3:] == ' 편역':
        author = author[:-3]
    if author[-2:] == ' 저':
        author = author[:-2]
    if author[-2:] == ' 글':
        author = author[:-2]
    if author[-2:] == ' 편':
        author = author[:-2]
    if author[-2:] == ' 외':
        author = author[:-2]
    if author[-2:] == ' 역':
        author = author[:-2]
    #print(author)
    return author
```

```python
author = list(df_book[0]['출판사'].unique())     # 정기간행물은 저자가 없으니 출판사명으로 대신...
author += list(df_book[1]['저작자'].unique())
#print(author)

author = list(map(lambda a:trim_author(a), author))
#print(author)
```

```python
a_list = []
a_prefix = ['도서', '책', '간행물', '정기간행물']
a_postfix = ['', '검색', '명', '이름', '제목']

for a in a_prefix:
    for b in a_postfix:
        a_list.append(a + b)
        a_list.append(a + ' ' + b)
print(a_list)

d_list = []
d_prefix = ['가 쓴 ', '가 저자인 ', '이 쓴 ', '이 저자인 ', '꺼 ']

for a in d_prefix:
    for b in a_prefix:
        d_list.append(a + b)
print(d_list)

w_book = {
    'BOOKNAME': bookname,
    'AUTHOR': author,
    'A': a_list,
    'B': ['있어', '어디있어', '어디 있어', '있을까', '위치', '위치좀', '위치 좀', '찾아줘', ],
    'C': ['저자', '저자가', '글쓴이', '글쓴이가'],
    'D': d_list
}
t_book = [
    '{A} {BOOKNAME}',
    '{BOOKNAME} {A}',
    '{A} {BOOKNAME} {B}',
    '{BOOKNAME} {A} {B}',
    '{C} {AUTHOR}',
    '{AUTHOR} {D}',
    '{C} {AUTHOR} {A}',
    '{C} {AUTHOR} {A} {B}',
    '{AUTHOR} {D} {A}',
    '{AUTHOR} {D} {B}',
]
```

```python
import re

# return chunk key list
# get template pattern string
def get_key_from_template(str):
  ret = re.findall(r'\{.*?\}', str)
  return list(map(lambda x:x.replace('{', '').replace('}', ''), ret))
  # return ret

def add_samples(permute):
    samples = []
    for chunks in permute:
      cur_q = template
      for i in range(len(chunks)):
        key = '{'+keylist[i]+'}'
        try:
          cur_q = cur_q.replace(key, chunks[i])
        except:
          continue
        #print(key, chunks[i], cur_q)

      samples.append(cur_q)

    return samples
```

```python
from itertools import product
import random

samples = []

for template in t_book:
  keylist = get_key_from_template(template)
      
  if len(keylist)==2:
    # Get all possible cases
    permute = list(product(w_book[keylist[0]],w_book[keylist[1]]))
  elif len(keylist)==3:
    # Get all possible cases
    permute = list(product(w_book[keylist[0]],w_book[keylist[1]],w_book[keylist[2]]))
  elif len(keylist)==4:
    # Get all possible cases
    permute = list(product(w_book[keylist[0]],w_book[keylist[1]],w_book[keylist[2]],w_book[keylist[3]]))
  
  # Random sampling
  permute = random.sample(permute, 1000)

  samples += add_samples(permute)
```
