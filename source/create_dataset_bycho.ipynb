```python
a_list = []
for i in range(1, 40):
    a_list.append(str(i) + '호봉')
#print(a_list)

b_list = []
for i in range(1962, 2023):
    b_list.append(str(i) + '년')
    b_list.append(str(i) + '년도')
    b_list.append(str(i) + '연도')
    b_list.append(str(i) + '년에')
    b_list.append(str(i) + '년도에')
    b_list.append(str(i) + '연도에')
for i in range(1, 60):
    b_list.append(str(i) + '년전')
    b_list.append(str(i) + '년전에')
    b_list.append(str(i) + '년 전')
    b_list.append(str(i) + '년 전에')
#print(b_list)

w_pay = {
    'CLASS': classes,
    'A': a_list + ['', '되면', '은', '는', '진급시', '진급 시', '진급하면' '진급되면', '처음', '처음에', '첫'],    # 호봉
    'B': b_list + ['요즘', '현재', '지금'],    # 연도 등 시기
    'C': ['월급', '월급이', '급여', '급여가', '봉급', '봉급이', '수입', '수입이', '연봉', '연봉이'],
    'D': ['', '궁금', '궁금쓰', '궁금함', '궁금해', '말해', '말해줘', '알고싶어', '알려줘', '어느정도', '어느 정도', '어느정도야', '어느 정도야', '어느정도임', '어느 정도임', '얼마', '얼마나', '얼마야', '얼마임', '얼마정도', '얼마 정도']
}
t_pay = [
    '{CLASS} {C} {D}',
    '{CLASS} {A} {C} {D}',
    '{CLASS} {B} {C} {D}',
]
```

```python
import re

# return chunk key list
# get template pattern string
def get_key_from_template(str):
  ret = re.findall(r'\{.*?\}', str)
  return list(map(lambda x:x.replace('{', '').replace('}', ''), ret))
  # return ret

def add_samples(permute):
    samples = []
    for chunks in permute:
      cur_q = template
      for i in range(len(chunks)):
        # print(key, chunks[i], cur_q)
        key = '{'+keylist[i]+'}'
        cur_q = cur_q.replace(key, chunks[i])

      samples.append(cur_q)

    return samples
```

```python
from itertools import product
import random

samples = []

for template in t_pay:
  keylist = get_key_from_template(template)
      
  if len(keylist)==3:
    # Get all possible cases
    permute = list(product(w_pay[keylist[0]],w_pay[keylist[1]],w_pay[keylist[2]]))
  elif len(keylist)==4:
    # Get all possible cases
    permute = list(product(w_pay[keylist[0]],w_pay[keylist[1]],w_pay[keylist[2]],w_pay[keylist[3]]))

    # Random sampling
    permute = random.sample(permute, 1000)

  samples += add_samples(permute)
```

```python
df_pay = pd.DataFrame(samples, columns=['question'])
df_pay
```
