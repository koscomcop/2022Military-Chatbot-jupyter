#### 개체추출 ==> 지역

```python
def get_dfs(file_names):
    dfs = []
    for fnm in file_names:
        sample_df = pd.read_csv(FILES_DIR+fnm, encoding='utf-8')
        dfs.append(sample_df)
        print(f'[{fnm.split(".")[0]}] # of data : {len(sample_df.index)} ======================================>')
        display(sample_df.head())
        print('\n')
        
    return dfs
```

```python
df_housing = get_dfs(file_names)
```

```python
# 지역 카테고리 따로 있는경우
locales = list(df_housing[0]['TMO명'].unique())
locales += list(df_housing[2]['지역1'].unique())
locales += list(df_housing[2]['지역2'].unique())
locales += list(df_housing[4]['지역'].unique())
print(set(locales))
```

```python
# 주소정보에서 지역 추출 : df_housing[1]
df_housing1_locales = list(df_housing[1]['주소'].apply(lambda x:x.split(' ')[0]).unique())
print(df_housing1_locales)
locales += df_housing1_locales
```

```python
# 주소정보에서 지역 추출 : df_housing[3]
df_housing3_locales = list(df_housing[3]['위치 주소'].apply(lambda x:x.split(' ')[0]).unique())
print(df_housing3_locales)
locales += df_housing3_locales
```

```python
locales = list(set(locales))
print(locales)
```

#### 개체추출 : 시설

```python
facilities = ['TMO', '여행장병 안내소', '여행장병안내소']
```

```python
# 시설명 따로 있는경우
facilities += list(df_housing[1]['시설명'].unique())
facilities += list(df_housing[2]['시설명'].unique())
facilities += list(df_housing[3]['시설 명'].unique())
facilities += list(df_housing[4]['시설명'].unique())
facilities = list(set(facilities))
print(facilities)
```

```python
# 시설명에서 불용어, 특수기호 제거
stw_f = ['(국.복지정책과)', '(수송사)', '(민간위탁)']
for i, x in enumerate(facilities):
    for s in stw_f:
        facilities[i] = facilities[i].replace(s, '')
    loc_idx = facilities[i].find('(')
    if loc_idx > 0:
        facilities[i] = facilities[i][:loc_idx]
    
print(set(facilities))
```
